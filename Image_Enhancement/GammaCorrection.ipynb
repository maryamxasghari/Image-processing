```python
import math
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

a = Image.open('../images/lena_color.bmp').convert('L')
b = np.asarray(a)

gamma = 0.5

b1 =  b.astype(float)
b3 = np.max(b1)

b2 = b1/b3

b3 = np.log(b2)*gamma
c = np.exp(b3)*255.0

d = Image.fromarray(c)
d.show()
```
