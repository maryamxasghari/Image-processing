````python
import matplotlib.pyplot as plt
````

````python
from PIL import Image

# open an image
img = Image.open('./images/profile.jpg')
img.show()
````

````python

# gray scale 
gray = Image.open('./images/profile.jpg').convert('L')
gray.show()
````

````python
import os

#convert the image format
filelist=['images/profile.jpg','images/moon.jpg']


for infile  in filelist:
    outfile = os.path.splitext(infile)[0]+".png"
    if infile !=outfile:
        try:
            Image.open(infile).save(outfile)
        except IOError:
                print("Cannot convert"), infile
````

````python
filelist=['images/girlface.bmp','images/houses.bmp']


for infile  in filelist:
    outfile = os.path.splitext(infile)[0]+".jpg"
    if infile !=outfile:
        try:
            Image.open(infile).save(outfile)
        except IOError:
                print("Cannot convert"), infile
````

````python
#Resize
smaller_img = img.resize((150,150))
# smaller_img.show()
plt.imshow(smaller_img)
````

````python
#Crop
box = (100,100,400,400)
region = img.crop(box)

# region.show()
plt.imshow(region)
````

````python
#Rotation
rotated_img = img.rotate(45)
# rotated_img.show()
plt.imshow(rotated_img)
````

````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from pylab import *

from skimage import data
scanned =  data.page()

plt.imshow(scanned)
plt.show()
````

````python
plt.imshow(scanned, cmap =cm.gray)
plt.show()
````

````python
thres =  np.zeros(shape(scanned)).astype('uint8')
threshold = 150
thres[scanned <threshold] =0
thres[scanned >=threshold] = 255

plt.imshow(thres, cmap=cm.gray)

plt.show()
````
